# **Bibliotecas**

```python
import numpy as np
import pandas as pd
import sklearn.metrics as skm
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor as rfs
from sklearn.neural_network import MLPRegressor 
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression 
from mlxtend.feature_selection import SequentialFeatureSelector as sfs
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import RandomizedSearchCV
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs

from scipy.stats import uniform as sp_randFloat
from scipy.stats import randint as sp_randInt

def mape(Y_actual,Y_Predicted):
    mape = np.mean(np.abs((Y_actual - Y_Predicted)/Y_actual))*100
    return mape
```

```python
dados = pd.read_csv('et_cfsr.csv')
dados_validation = pd.read_csv('et_power.csv')

dados.head(), dados_validation.head()
```

# **Pré processamento**

```python
x1 = dados[['lon','lat','year','month']]
x2 = dados[['prec','temp']]
y = dados['etp']

x1_vali= dados_validation[['lon','lat','year','month']]
x2_validation = dados_validation[['prec','temp']]
y_validation = dados_validation['etp']
```

```python
x2 = StandardScaler().fit_transform(x2)
x2 = pd.DataFrame(x2, columns=[dados.columns[5:7]])
x2.head()

x2_validation = StandardScaler().fit_transform(x2_validation)
x2_validation = pd.DataFrame(x2_validation, columns=[dados_validation.columns[5:7]])
x2_validation.head(),x2_validation.head()
```

```python
x = pd.concat([x1,x2],axis=1, join='inner')
x_validation = pd.concat([x1_vali,x2_validation],axis=1, join='inner')

x.head(),x_validation.head()
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=.3, random_state=1)
```

# **Random Forest**

```python
rf_modelo = rfs().fit(x_treino, y_treino)
             
```

## **Random Search param**

```python
hyperpar_space = {'max_depth':[3,5,7,9],
                  'n_estimators':[50,100,150,200,500],
                  'max_features':[2,4,6,8]
                 }

rs_rf = RandomizedSearchCV(rf_modelo, param_distributions=hyperpar_space,
                        n_iter=20, scoring="r2", random_state=1,
                        n_jobs=-1, cv=10, return_train_score=True).fit(x_treino,y_treino)
```

```python
hyperpar_space
```

```python
pd.DataFrame.from_dict(rs_rf.cv_results_).to_csv('random_Forest_iter.csv')
pd.DataFrame.from_dict(rs_rf.cv_results_)
```

```python
pd.DataFrame.from_dict(rs_rf.cv_results_)['mean_train_score'].plot()
```

```python
print("Optimal hyperparameter combination:", rs_rf.best_params_)
print()
print("Mean cross-validated training accuracy score:",
      rs_rf.best_score_)
print()
```

### ***Feature Importance***

```python
importancia = rs_rf.best_estimator_.feature_importances_
importancia_graf = pd.DataFrame({'lon':[importancia[0]],'lat':[importancia[1]], 
                                 'year':[importancia[2]],'month':[importancia[3]],
                                 'prec':[importancia[4]], 'temp':[importancia[5]]
                                })
importancia_graf
```

```python
sns.barplot(data=importancia_graf, orient='h');
```

## **Treino**

```python
yest_treino = rs_rf.best_estimator_.predict(x_treino)
```

```python
dadosgraf = pd.DataFrame({'Yobs':y_treino,
                         'Yest': yest_treino})
dadosgraf.head()
```

```python
sns.lmplot(x='Yobs',y='Yest',data=dadosgraf);
```

```python
r2 = skm.r2_score(y_treino,yest_treino)
mse=skm.mean_squared_error(y_treino,yest_treino)
rmse = mse**0.5
mae = skm.mean_absolute_error(y_treino, yest_treino)
rrmse = rmse*100/np.mean(y_treino)
rf_mape = mape(y_treino,yest_treino)

r2,rmse,rrmse,rf_mape
```

```python
x=[30,150]
y=x

points = plt.scatter(y='Yest', x='Yobs',data=dadosgraf,
                     c=dadosgraf["Yobs"],marker="8", s=80, cmap="viridis")  #cmap é a cor da barra  
cbar = plt.colorbar(points, shrink=1)
#cbar.set_label('XCO$\mathregular{_{2}}$ estimated',fontsize=14, family='Segoe UI Bold')  #COLOCAR O NOME DA BARRA
sns.regplot(y='Yest', x='Yobs', data=dadosgraf, scatter=False, color=".1")

plt.plot(x,y, 'b--',data=None)

pvalue = 0.001
plt.text(40, 140, 'R$\mathregular{^{2}}$ ETP estimated= %0.4f' % r2,family='Segoe UI Bold',fontsize=8, color= 'black')
plt.text(40,136, 'RMSE = %0.2f'% rmse,family='Segoe UI Bold',fontsize=8)
plt.text(40,132, 'RRMSE = %0.2f'% rrmse,family='Segoe UI Bold',fontsize=8)
plt.text(40,128, 'MAPE = %0.2f'% rf_mape,family='Segoe UI Bold',fontsize=8)
plt.text(40, 124,'pvalue <%0.3f' % pvalue,family='Segoe UI Bold',fontsize=8)
plt.xlabel('ETP observed (mm month$\mathregular{^{-1}}$)',fontsize=14, family='Segoe UI Bold')
plt.ylabel('ETP estimated (mm month$\mathregular{^{-1}}$)',fontsize=14, family='Segoe UI Bold')
plt.rcParams['figure.figsize'] = (4,6)

#plt.savefig('.png', transparent = False,bbox_inches='tight', dpi=1000)
```

```python
ann_modelo= MLPRegressor().fit(x_treino,y_treino)
```

## **Random Search param**

```python
hyperpar_space_mlp = {'hidden_layer_sizes':[(5,5),(10,10),(10,5,10),(10,10,10)],
                      'activation':['logistic','tanh','identity'],
                      'max_iter':sp_randInt(1, 1000),
                      'learning_rate_init': sp_randFloat(),
                      'solver':['adam','sgd','lbfgs'],
                      'learning_rate':['constant', 'invscaling', 'adaptive'],
                      'alpha':sp_randFloat()
                     }
```

```python
rs_mlp = RandomizedSearchCV(ann_modelo, param_distributions=hyperpar_space_mlp,
                        n_iter=20, scoring="r2", random_state=1,
                        n_jobs=-1, cv=10, return_train_score=True).fit(x_treino,y_treino)
```

```python
print("Optimal hyperparameter combination:", rs_mlp.best_params_)
print()
print("Mean cross-validated training accuracy score:",
      rs_mlp.best_score_)
```
